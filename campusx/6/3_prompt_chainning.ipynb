```python
from langgraph.graph import StateGraph,START,END

from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
model=ChatOpenAI()
```

```python
class Blog(TypedDict):
    score:int
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: Blog) -> Blog:
    prompt=f"Create an outline for a blog titled '{state['title']}'"
    outline=model.invoke(prompt).content
    state['outline']=outline
    return state
```

```python
def create_content(state: Blog) -> Blog:
    prompt=f"Create a detailed blog post based on the following outline:\n{state['outline']}"
    content=model.invoke(prompt).content
    state['content']=content
    return state
```

```python
def calculate_score(state: Blog) -> Blog:
    prompt=f"Evaluate the quality of the following blog post and provide a score out of 10:\n{state['content']}"
    score=model.invoke(prompt).content
    state['score']=score
    return state
```

```python
graph=StateGraph(Blog)
# node
graph.add_node("create_outline",create_outline)
graph.add_node("create_content",create_content)
graph.add_node("calculate_score",calculate_score)
# edges
graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","create_content")
graph.add_edge("create_content","calculate_score")
graph.add_edge("calculate_score",END)

workflow=graph.compile()
```

```python
initial_state={"title":"The Future of AI in Everyday Life"}
final_state=workflow.invoke(initial_state)
```
